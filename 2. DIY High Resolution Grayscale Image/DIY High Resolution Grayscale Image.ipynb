### 1. Lets Create high pixel density images

```python
img  = []
for i in range(720):
    temp = []
    for j in range(1280):
        temp.append(0)
    img.append(temp)
img = np.array(img)        
```

```python
cv.imwrite('Gray_0.png', img )
```

```python
lst =[]
for i in range(256):
    temp = []
    for j in range(256):
        temp.append(i)
    lst.append(temp)
lst = np.array(lst)        
```

```python
cv.imwrite('Shades.jpg', lst)
```
