#### 1. Loading an Image using OpenCV

```python
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('../rgb.jpeg')

if img is None:
    print("Image not found")
else:
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

### 2. DIsplaying 3 Frames on RGB

```python
img = cv.imread('../rgb.jpeg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

r = img[ :  , : , 0 ]  # - Red
g = img[ :  , : , 1 ]  # - Green
b = img[ :  , : , 2 ]  # - Blue


cv.imshow("Frame",r)
cv.waitKey(0)

cv.imshow("Frame",g)
cv.waitKey(0)

cv.imshow("Frame",b)
cv.waitKey(0)
```
