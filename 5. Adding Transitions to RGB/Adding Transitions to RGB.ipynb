```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Creating Different Shared of Red

```python
arr = np.array([[[ 30  , 0 , 0 ] , [ 60  , 0 , 0 ] ,[ 90  , 0 , 0 ]],
                [[ 120 , 0 , 0 ] , [ 150 , 0 , 0 ] ,[ 180 , 0 , 0 ]],
                [[ 210 , 0 , 0 ] , [ 240 , 0 , 0 ] ,[ 255 , 0 , 0 ]]])

plt.imshow(arr)
```

#### 2. Creating Different Shades of Green

```python
arr = np.array([[[ 0, 30  , 0  ] , [ 0, 60  , 0 ] ,[ 0, 90  , 0 ]],
                [[ 0, 120 , 0  ] , [ 0, 150 , 0 ] ,[ 0, 180 , 0 ]],
                [[ 0, 210 , 0  ] , [ 0, 240 , 0 ] ,[ 0, 255 , 0 ]]])

plt.imshow(arr)
```

#### 5. Creating Shades of Red - Dense

```python
img = []

for i in range(256):
    t = []
    for j in range(256):
        t.append([i,0,0])
    img.append(t)
img = np.array(img)
plt.imshow(img)        
```

#### 6. Creating Shades of Green - Dense

```python
img = []
for i in range(256):
    t  = []
    for j in range(256):
        t.append([0,i,0])
    img.append(t)
img = np.array(img)  
plt.imshow(img)      
```

#### 7. Creating Shades of Blue - Dense

```python
img = []

for i in range(256):
    t = []
    for j in range(256):
        t.append([0,0,i])
    img.append(t)

img = np.array(img)

plt.imshow(img)
```

#### 7. Merging these three shades

```python
img = []

# Red

for i in range(256):             # from Black to Red
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)

for i in range(256,0,-1):        # from Red to Black
    t = []
    for j in range(1536):
        t.append([i,0,0])
    img.append(t)
    
# Green
    
for i in range(256):             # from Black to Green
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

for i in range(256,0,-1):        # from Green to Black
    t = []
    for j in range(1536):
        t.append([0,i,0])
    img.append(t)

#Blue    


for i in range(256):             # from Black to Blue
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)

for i in range(256,0,-1):        # from Blue to Black
    t = []
    for j in range(1536):
        t.append([0,0,i])
    img.append(t)
    
    

img = np.array(img, dtype=np.uint8)

plt.imshow(img)
```

```python
cv.imwrite('Transition.jpg',img)
```
