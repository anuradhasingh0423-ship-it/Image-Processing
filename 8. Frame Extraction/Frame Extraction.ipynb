```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

#### 1. Loading an Image using OpenCV

```python
img = cv.imread('../5. Adding Transitions to RGB/Transition.jpg')

img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

plt.imshow(img)
```

#### 2. Frame Extraction

```python
r = img[: , : , 0] # red
g = img[: , : , 1] #green
b = img[: , : , 2] #blue
```

```python
plt.imshow(r)
```

```python
plt.imshow(b)
```

```python
plt.imshow(g)
```

```python
cv.imwrite('Blue_t.jpg',b)
cv.imwrite('Red_t.jpg',r)
cv.imwrite('Green_t.jpg',g)
```
